```python

import numpy as np
import random
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
n= 250

melons_width = np.random.normal(35, 4, n) #برحسب سانتی متر
melons_height = np.random.normal(8, 2, n)# برحسب سانتی متر 
melons_weight = np.random.normal(10, 0.5, n)# برحسب کیلوگرم
dict_melon = {'wid':melons_width, 'hei':melons_height, 'wei':melons_weight}

balloon_width = np.random.normal(40, 4, n)
balloon_height = np.random.normal(10, 3, n)
balloon_weight = np.random.normal(0.5, 0.02, n)# برحسب کیلوگرم
dict_balloon = {'wid':balloon_width, 'hei':balloon_height, 'wei':balloon_weight}

ax = plt.axes(projection='3d')
ax.scatter(dict_melon['wid'], dict_melon['hei'], dict_melon['wei'],marker='^', c='orange', label='Melons')

ax.scatter(dict_balloon['wid'], dict_balloon['hei'], dict_balloon['wei'], c='Blue', label='Ballons')

ax.legend()

ax.set_xlabel('Width')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')

plt.show()
```

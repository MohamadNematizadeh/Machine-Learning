```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

load dateset

```python
data = load_boston()
bostons_data = pd.DataFrame(data.data, columns=data.feature_names)
bostons_data.head()
```

Preprocess dateset

```python
bostons_data['MEDV'] = data.target
x = bostons_data[['AGE', 'LSTAT']]
y = bostons_data[['MEDV']]
y.head()
```

```python
x = x.to_numpy()
y = y.to_numpy()
```

M = (X.T * X)^1  *  (X.T * Y)

```python
m = np.matmul(np.linalg.inv(np.matmul(x.T, x)), np.matmul(x.T, y))
m
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:,0], x[:,1], y)
ax.set_xlabel('AGE')
ax.set_ylabel('LSTAT')
ax.set_zlabel('House Price')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

x_arange = np.arange(x[:,0].min(), x[:,0].max())
y_arange = np.arange(x[:,1].min(), x[:,1].max())
px, py = np.meshgrid(x_arange, y_arange)
 
pz = px * m[0] + py * m[1]
ax.plot_surface(px, py, pz, alpha = 0.4)
ax.scatter(x[:,0], x[:,1], y)
ax.set_xlabel("AGE")
ax.set_ylabel("LSTAT")
ax.set_zlabel("House Price")
plt.show()
```

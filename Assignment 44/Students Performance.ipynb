```python
X = np.random.uniform(0, 75, 200)
Y = X + np.random.uniform(0, 25, 200) 
plt.scatter(X ,Y,c='green')
plt.show()
```

```python
y_pred = X * 1.5
plt.scatter(X, Y, label='original data',c='green')
plt.plot(X, y_pred, c='hotpink', lw=2, label='random fitted line')
plt.legend(loc='best')
plt.show()
```

```python
x = X.reshape((200,1))
y = Y.reshape((200,1))
m = np.matmul(np.linalg.inv(np.matmul(x.T, x)), np.matmul(x.T, y))
m = m[0,0]
Y_pred = x * m
plt.scatter(x, y, label='original data',c='green')
plt.plot(x, Y_pred, c='red', lw=2, label='fitted line')
plt.legend(loc='best')
plt.show()
```

```python
res = stats.linregress(X, Y)
plt.plot(x, y, 'o', label='original data',c='green')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line-scipy', lw=2)
plt.legend(loc='best')
plt.show()
```

```python
linregress = stats.linregress(X, Y)
plt.scatter(X ,Y, label='orginal data',c='green')
plt.plot(X , Y_pred, c = 'blue', lw = 3, label='linear least squares line')
plt.plot(X , linregress.intercept + linregress.slope * X, c = 'red', lw = 3, label='linregress line')
plt.legend(loc='best')
plt.show()
```

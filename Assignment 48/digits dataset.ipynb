```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from mlp import  Mlp
```

```python
dataset = load_digits()

X = dataset.data
Y = dataset.target
Y = np.eye(10)[Y]  # one hot
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)

X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
 #learning rate
D_in = X_train.shape[1]
H1 = 110
H2 = 64
D_out = Y_train.shape[1]

η = 0.001
epochs = 80
func1="sigmoid"
func2="softmax"
```

```python
mlp=Mlp(D_in,H1,H2,D_out,epochs,η,func1,func1,func2)

losses,accuracies=mlp.fit(X_train,Y_train)

loss_train,accuracy_train = mlp.evaluate( X_train,Y_train)
loss_test,accuracy_test = mlp.evaluate( X_test, Y_test)

print(f"loss_test: {loss_test}       accuracy_test: {accuracy_test}")
print(f"loss_train: {loss_train}       accuracy_train: {accuracy_train}")
```

```python
plt.figure(figsize=(10,6))
# loss
plt.subplot(1,2,1)
plt.plot(losses)
plt.ylabel("loss")
# accuracy
plt.subplot(1,2,2)
plt.plot(accuracies)
plt.ylabel("accuracy")
plt.show()
```

```python
import cv2

image = cv2.imread("input/1.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = image.astype(np.float32)
x = image
y_pred=mlp.predict(x)
print(f"result: {y_pred}")
```

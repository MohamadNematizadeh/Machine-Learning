```python
import os
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('input/mask.jpg', 0)
plt.imshow(img, cmap='gray')
```

```python
rows, cols = img.shape
numbers = []
for i in range(10):
    target_nums = img[int(rows * (i/10)): int(rows * ((i+1)/10)), :]
    numbers.append(target_nums)
counter = 0
for i in range(len(numbers)): 
    newpath = f"output/{i}"
    if not os.path.exists(newpath):
        os.makedirs(newpath)
    
    rows, cols = numbers[i].shape
    
    for j in range(5): # rows in target numbers img
        for k in range(100): # cols in target numbers img
            target_num = numbers[i][int(rows * (j/5)): int(rows * ((j+1)/5)), int(cols * (k/100)): int(cols * ((k+1)/100))]
            
            cv2.imwrite(f"{newpath}/{counter}.jpg", target_num)
            counter += 1
    counter = 0
```
